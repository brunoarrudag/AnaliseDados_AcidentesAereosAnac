```python
acidentes_aereos.where(acidentes_aereos["Classificacao_da_Ocorrencia"] == "Acidente", inplace=True)

acidentes_aereos
```

```python
contagem_valores = acidentes_aereos['Classificacao_da_Ocorrencia'].value_counts()
#Plotagem do gr√°fico de barras categorizar as quantidades de Acidentes e Incidentes.
contagem_valores.plot(kind='bar', figsize=(10, 6))
plt.title('Contagem de Acidentes Por categorias')
plt.xlabel('Valores')
plt.ylabel('Contagem')
plt.show()
```

```python
acidentes_aereos['Data_da_Ocorrencia'] = pd.to_datetime(acidentes_aereos['Data_da_Ocorrencia'],
format='%Y-%m-%d')

#Extrair o ano da coluna Data da Ocorrencia:
acidentes_aereos['ano'] = acidentes_aereos['Data_da_Ocorrencia'].dt.year

# Agrupar os dados por ano e contar o n√∫mero de acidentes por ano
acidentes_por_ano = acidentes_aereos.groupby('ano').size()

# Gerar um gr√°fico de linha mostrando a evolu√ß√£o dos acidentes ao longo dos anos
acidentes_por_ano.plot(kind='line', figsize=(10, 6))
plt.title('Evolu√ß√£o dos Acidentes A√©reos ao Longo dos Anos')
plt.xlabel('Ano')
plt.ylabel('N√∫mero de Acidentes')
plt.grid(True)
plt.show()
```
